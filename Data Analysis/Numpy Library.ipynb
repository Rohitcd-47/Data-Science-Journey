# Deep copy and shallow copy

```python
# whenever you are dealing with array then any changes you made in array, array creates a new
# space for that in RAM,it does not affect the original array unlike list
a = np.arange(1,21)
a
```

```python
slice = a[:5]
slice = slice*5
slice
```
