# Valid sudoku

```python
s = np.array([
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],
    
    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],
    
    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
])
```

```python
c = np.sum(s,axis=1)
for i in c:
    if i != 45:
        print("Sudoku is not valid")
        break
else:
    print("For rows it is valid")
```

```python
r = np.sum(s,axis=0)
for i in r:
    if i != 45:
        print("Sudoku is not valid")
        break
else:
    print("For columns it is valid")
```

```python
for i in range(0,9,3):
    for j in range(0,9,3):
        n = s[i:i+3,j:j+3]
        print(n.sum())
```

# General Questions

```python
# Columns: [Age, Math Marks, Science Marks]
data = np.array([
    [18, 85, 78],   # Student 1
    [19, 92, 88],   # Student 2
    [17, 76, 95],   # Student 3
    [18, 65, 70],   # Student 4
    [20, 90, 85]    # Student 5
])
```

```python
#Get the shape of the matrix.
data.shape
```

```python
#Find the average age of students.
slice = data[:,0:1]
avg_age = round(np.mean(slice))
print("The average age of students :",avg_age)
```

```python
#Extract Math marks of all students.

math_marks = data[:,1:2]
math_marks
```

```python
#Find the highest Science mark.

round(np.max(data[:2]))
```

```python
#Get details of the student who scored more than 90 in Math.

data[data[:,1]>90]
```
