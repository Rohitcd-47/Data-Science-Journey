```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Matplotlib Basics**

```python
x = np.linspace(0,5,11)
y= x**2
```

```python
plt.plot(x,y)
plt.title("Simple Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

**Subplots in Matplotlib**

```python
plt.subplot(2,2,1)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.subplot(2,2,3)
plt.subplot(2,2,4)
plt.plot(x,y)
```

**Object oriented way**

```python
fig = plt.figure(figsize=(10,8),dpi=100)
# 0th index : left adjusment percentage 0.1 = 10%
# 1st index : Right adjusment percentage 0.1 = 10%
# 2nd index : Width adjusment percentage 0.5 = 50%
# 3rd index : Height adjusment percentage 0.1 = 10%

axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.4,0.4,0.5,0.6])
axis2.plot(x,y)
```

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.
figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch).

**Types of Plots**

**Saving plots**

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("basicplot.png")
```
