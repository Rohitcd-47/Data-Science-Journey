# Dictionary Constructor

```python
a = dict(name="Rohit",age=22,country="India")
print(a)
```

```python
# change value
a['age'] = 23
print(a)
```

```python
# add value
a.update({"maid id" : "rohit@gmail.com"})
print(a)
```

```python
# remove
a.pop('maid id')
print(a)
```

```python
# if you want to clear all key-values pairs in dictionary
a.clear()
print(a)
```

# Unit-11 Sets

```python
# agar hume different data types ke elements ek sath store karna hein toh hum set use karenge
# list and tuple mein order elements hote hein
# set mein unordered hote hein and duplicates allow nhi hein

# creating set

a = {'animals',12,'ball','cat',12,True,1,False,0}
print(type(a))
print(a)
```

```python
# here true and 1  & false and 0 are same
```

```python
print(len(a))
```

```python
# you cannot write set inside set like list and tuple
# creating set using constructor

a = set((1,3,5,7))
print(a)
print(type(a))
```

```python
# empty set is dictionary

a = {}
print(type(a))

b = set()
print(type(b))
```

```python
# join 2 sets

x = {10,20,30}
y = {'a','b','c'}
z = x.union(y)
print(z)
```
